```python
#neural network 
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import sklearn.model_selection as sk
from sklearn.model_selection import train_test_split
```

```python
X=np.array([1.,2.,3.,4.,5.,6.,7.,8.])
y=np.array([2.,4.,6.,8.,10.,12.,14.,16.])
X,y
```

```python
X=tf.cast(tf.constant(X),dtype=tf.float32)
y=tf.cast(tf.constant(y),dtype=tf.float32)
X,y
```

```python
tf.random.set_seed(42) #why??

model=tf.keras.Sequential([tf.keras.layers.Dense(50,activation='relu'),tf.keras.layers.Dense(1)])
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),
            metrics=['mae'])
```

```python
model.fit(X,y,epochs=100,verbose=2)
```

```python



model.predict([10.])
```

```python
# lets make a bigger dataset, ready lets go
X=tf.range(-100,100,4)
# X=tf.cast(X,dtype=tf.float32)
X.shape
```

```python
y=X+10
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,
                                                    random_state=42)
```

```python
X_train=X[:40]
y_train=y[:40]
X_test=X[40:]
y_test=y[40:]
```

```python
plt.figure(figsize=(10,7))
```

```python
tf.random.set_seed(42)
model=tf.keras.Sequential([tf.keras.layers.Dense(10,input_shape=[1], name="input_layer"),tf.keras.layers.Dense(1,name="output_layer")],name='mode1_1')
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(0.01),
              metrics=['mae'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100,verbose=0)
```

```python
#visulize it our perdiction 
y_pred=model.predict(X_test)

y_pred
```

```python
y_test
```

```python
#lets creat a plotting function 
def plot_predictions(train_data=X_train, 
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                    pred=y_pred):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data,train_labels,c='b',label='training data')
  plt.scatter(test_data,test_labels,c='g',label='testing data')
  plt.scatter(test_data,pred,c='r',label='predictions')
  plt.legend(); 
  plt.show();
```

```python
plot_predictions()
```

```python
model.evaluate(X_test,y_test)
```
